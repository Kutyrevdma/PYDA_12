```python
# Задание 1. Какая фраза длиньше

"""

Пользователь вводит 2 переменных с клавиатуры.
Программа проверяет какая строка больше или строки равны.

Например:

>>>Введите строку для переменной phrase_1:В этой строке 25 символов
>>>Введите строку для переменной phrase_2:В этой строке 26 символов.

phrase_2 длинее phrase_1

"""

phrase_1 = len(input('Введите строку для переменной phrase_1: '))
phrase_2 = len(input('Введите строку для переменной phrase_2: '))

if phrase_1 > phrase_2:
    print('phrase_1 длинее phrase_2')

elif phrase_1 < phrase_2:
    print('phrase_2 длинее phrase_1')

else:
    print('phrase_1 и phrase_2 равны')
```

```python
# Задание 2. Високосный код

"""

Пользователь вводит год с клавиатуры.
Программа проверяет високосный или обычной год

Например:

>>>Введите год: 1524
Високостный

"""
year = int(input('Введите год: '))

if year % 4 != 0:
    print('Обычный год')

elif year % 100 == 0:
    if year % 400 == 0:
        print('Високостный')
    else:
        print('Обычный')
else:
    print('Високостный')
```

```python
# Задание 3. Знак зодиака

"""

Пользователь вводит день и месяц с клавиатуры.
Программа выводит знак зодиака

Например:

>>>Введите день: 18
>>>Введите месяц: Март
Рыбы

"""
day = int(input('Введите день: '))
month = input('Введите месяц: ').capitalize()

if 0 < day >= 31:
    print('В месяце 31 день')

elif ((month == 'Январь') and (day >= 20)) or \
        ((month == 'Февраль') and (day <= 18)):
    print('Водолей')

elif ((month == 'Февраль') and (day >= 19)) or \
        ((month == 'Март') and (day <= 20)):
    print('Рыбы')

elif ((month == 'Март') and (day >= 21)) or \
        ((month == 'Апрель') and (day <= 20)):
    print('Овен')

elif ((month == 'Апрель') and (day >= 21)) or \
        ((month == 'Май') and (day <= 20)):
    print('Телец')

elif ((month == 'Май') and (day >= 21)) or \
        ((month == 'Июнь') and (day <= 21)):
    print('Близнецы')

elif ((month == 'Июнь') and (day >= 22)) or \
        ((month == 'Июль') and (day <= 22)):
    print('Рак')

elif ((month == 'Июль') and (day >= 23)) or \
        ((month == 'Август') and (day <= 23)):
    print('Лев')

elif ((month == 'Август') and (day >= 24)) or \
        ((month == 'Сентябрь') and (day <= 23)):
    print('Дева')

elif ((month == 'Сентябрь') and (day >= 24)) or \
        ((month == 'Октябрь') and (day <= 23)):
    print('Весы')

elif ((month == 'Октябрь') and (day >= 24)) or \
        ((month == 'Ноябрь') and (day <= 22)):
    print('Скорпион')

elif ((month == 'Ноябрь') and (day >= 23)) or \
        ((month == 'Декабрь') and (day <= 21)):
    print('Стрелец')

elif ((month == 'Декабрь') and (day >= 22)) or \
        ((month == 'Январь') and (day <= 20)):
    print('Козерог')
    
else:
    print('Вы указали неверно месяц')
```

```python
# Задание 4. Подбираем коробку

"""

Пользователь вводит ширину, длину и высоту коробки.
Программа выводит необходимую коробку для товара

Например:

>>>Введите width: 18
>>>Введите length: 220
>>>Введите height: 18
Упаковка для лыж

"""
width = int(input('Введиты width: '))
length = int(input('Введиты length: '))
height = int(input('Введиты height: '))

if (width < 15) and (length < 15) and (height < 15):
    print('Коробка № 1')

elif length > 200:
    print('Упаковка для лыж')

elif (50 > width > 15) or (50 > length > 15) or (50 > height > 15):
    print('Коробка № 2')

else: 
    print('Стандартная упакова')
```

```python
# Задание 5. Счаслитвый билет

"""

Пользователь вводит номер билета.
Программа выводит результат счастливый билет или нет

Например:

>>> 111111
Счастливый билет

"""
ticket = int(input('Введите номер билета: '))

ticket_number_1 = ticket // 100000
ticket_number_2 = (ticket % 100000) // 10000
ticket_number_3 = (ticket % 10000) // 1000
ticket_number_4 = (ticket % 1000) // 100
ticket_number_5 = (ticket % 100) // 10
ticket_number_6 = (ticket % 10) // 1

if (ticket_number_1 + ticket_number_2 + ticket_number_3) == (ticket_number_4 + ticket_number_5 + ticket_number_6):
    print('Счастливый билет')
else:
    print('Обычный билет')
```

```python
# Задание 6. Площаь фигуры

"""

Прользователь вводит название фигуры.
Программа выводит площадь фигуры

Например:

>>>Введите название фигуры (Доступные: круг, прямоугольник, треугольник, квадрат): треугольник
>>>Прямоугольный или равносторонний ?равносторонний
>>>Введите сторону A: 10
Площадь Равностороннего Треугольника: 43.30127018922193

"""

figure = input('Введите название фигуры (Доступные: круг, прямоугольник, треугольник, квадрат): ').capitalize()

# Круг

if figure == 'Круг':
    circle_radius = int(input('Введите радиус круга: '))
    circle_area = 3.14 * (circle_radius ** 2)
    print(f'Площадь {figure}а: {circle_area}')

# Прямоугольник

elif figure == 'Прямоугольник':
    rectangle_length = int(input('Введите длину: '))
    rectangle_width = int(input('Введите ширину: '))
    rectangle_area = rectangle_length * rectangle_width
    print(f'Площадь {figure}а: {rectangle_area}')
    
# Квадрат

elif figure == 'Квадрат':
    square_side = int(input('Введите сторону: '))
    square_area = square_side ** 2
    print(f'Площадь {figure}а: {square_area}')
    
# Треугольник

else:
    triangle_type = input('Прямоугольный или равносторонний ?').capitalize()
    
# Прямоугольный треугольник

    if triangle_type == 'Прямоугольный':
        triangle_side_A = int(input('Введите сторону А: '))
        triangle_side_B = int(input('Введите сторону В: '))
        triangle_area = 0.5 * (triangle_side_A * triangle_side_B)
        print(f'Площадь Прямоугольного {figure}а: {triangle_area}')
        
# Равносторонний треугольник 

    else:
        triangle_side_A = int(input('Введите сторону A: '))
        triangle_area = ((3 ** 0.5) / 4) * (triangle_side_A ** 2)
        print(f'Площадь Равностороннего {figure}а: {triangle_area}')                
```
