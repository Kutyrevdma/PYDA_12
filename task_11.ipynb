```python
%%time

# Задание № 1. Функция классифицирующая фильмы по следующим правилам:
# оценка 2 и меньше - низкий рейтинг
# оценка 4 и меньше - средний рейтинг
# оценка 4.5 и 5 - высокий рейтинг

    
def class_movies(row):
    # Функция, которая выводить рейтинг фильма:
    # низкий, средний, высокий
    if row['rating'] <= 2.0:
        return 'низкий рейтинг'
    elif 2.0 < row['rating'] <= 4.0:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'


ratings = pd.read_csv('./task_11_data/ratings.csv')[['movieId', 'rating']]

ratings['class'] = ratings.apply(class_movies, axis=1)
ratings.head()
```

```python
%%time
# Задание 2. Необходимо написать гео-классификатор, 
# который каждой строке сможет выставить географическую принадлежность определенному региону.

# Исходные данные.

geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

data = pd.read_csv('./task_11_data/keywords.csv')

# Создадим список городов.

city = []
for cities in geo_data.values():
    for citys in cities:
        city.append(citys)

# Создаем функцию по поиску ключевых слов.

def region(row):
    data = row['keyword'].split(' ')
    for word in data:
        if word in city:
            for items in geo_data.items():
                if word in items[1]:
                    return items[0]
    else:
        return 'undefined'

data['region'] = data.apply(region, axis=1)

data[data['region'] != 'undefined']
```
