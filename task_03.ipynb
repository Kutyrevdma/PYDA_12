```python
# Задание 1. Оставить уникальные значения.

"""

Пользователь вводит данные данные в виде хеш-таблицы (словаря): {пользователь: [гео_метка_1, ...гео_метка_N]}.
Программа считает и показывает уникальные значение по всем пользователям.

ids = {'user1': [213, 213, 213, 15, 213], 
       'user2': [54, 54, 119, 119, 119], 
       'user3': [213, 98, 98, 35]}
       
Результат: {98, 35, 15, 213, 54, 119}

"""

# Введенные данные пользователя.

ids = {'user1': [213, 213, 213, 15, 213], 
       'user2': [54, 54, 119, 119, 119], 
       'user3': [213, 98, 98, 35]}

# Объявление переменных.

item_set = set()
item_list = list()

# Получение уникальные пользователей через set()

for geo_tegs in ids.values():
    for geo_teg in geo_tegs:
        item_set.add(geo_teg)
print(f'Решение через set(): {item_set}')

# Получение уникальные пользователей через list()

for geo_tegs in ids.values():
    for geo_teg in geo_tegs:
        if geo_teg not in item_list:
            item_list.append(geo_teg)
            
# Вывод результата.

print(f'Решение через list(): {item_list}')
```

```python
# Задание 2. Распределение количества слов.

"""

Пользователь вводит список поисковых запросов пользователя 
Программа считает и показывает распределние количества слов.

queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
]

Поисковых запросов, содержащих 2 слов(а): 66.67%
Поисковых запросов, содержащих 3 слов(а): 33.33%

"""

# Введенные данные пользователя.

queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

# Объявление переменных.

item = dict()
sum_words = 0

# Подсчет результатов.

for words in queries:
    count_word_in_words = len(words.split(' '))
    if count_word_in_words not in item.keys():
        item.setdefault(count_word_in_words, 1)
    else:
        item[count_word_in_words] += 1  
    sum_words += 1
"""
Если переменная words будет пустой, то этот счётчик будет выводить не коректные данные...
Думал, что смогу уйти от этого так: queries.isalnum()...Понял, что пробел между слов дает False (Что логично)
Еще немного подумал и решил добавить вот такое условие в начале итерации:

for words in queries:
    word_strip = words.strip()
    if len(word_strip) == False:
        continue
    count_word_in_words = len(words.split(' '))
...

Кажется я изобретаю велосипед... Скажите, как-то можно еще уйти от этого ? 
"""
# Вывод результата.

print(f'Поисковых запросов, содержащих 2 слов(а): {round((100 * item[2]) / sum_words, 2)}%')
print(f'Поисковых запросов, содержащих 3 слов(а): {round((100 * item[3]) / sum_words, 2)}%')
```

```python
# Задание 3. Добовляем показатель ROI (ROI — это показатель возврата инвестиций).

"""

Пользователь вводит список компании о затратах и доходе рекламных кампаний.
Программа считает и показывает ROI.

results = {
'vk': {'revenue': 103, 'cost': 98},
'yandex': {'revenue': 179, 'cost': 153},

'vk': {'ROI': 5.1, 'cost': 98, 'revenue': 103},
'yandex': {'ROI': 16.99, 'cost': 153, 'revenue': 179}}

"""
# Импортируем модуль pprint для удобного вывода и чтения.

from pprint import pprint 

# Введенные данные пользователя.

results = {
'vk': {'revenue': 103, 'cost': 98},
'yandex': {'revenue': 179, 'cost': 153},
'facebook': {'revenue': 103, 'cost': 110},
'adwords': {'revenue': 35, 'cost': 34},
'twitter': {'revenue': 11, 'cost': 24},
}

# Подсчет результатов.

for data in results.values():
    roi = round(((data['revenue'] / data['cost']) - 1) * 100, 2)
    data['ROI'] = roi

# Вывод результата.

pprint(results)
```

```python
# Задание 4. 

"""

Пользователь вводит данные компании с значеним.
Программа выводит компанию с наибольшим значением.

stats = {'sport': 150, 'news': 300, 'films': 75}

Результат: Максимальный объем продаж на рекламном канале: news

"""

# Введенные данные пользователя.

stats = {'facebook': 55, 'yandex': 115, 'vk': 120, 'google': 99, 'email': 42, 'ok': 98}

# Объявление переменных.

max_value_1 = 0
max_value_2 = 0
max_company = str()

# Решение № 1.

for company, value in stats.items():
    if max_value_1 < value:
        max_value_1, max_company = value, company
print(f'Результат решения № 1: Максимальный объем продаж на рекламном канале: {max_company}')

# Решение № 2. 

result = [(value, company) for company, value in stats.items()] 
max_value_2 = max((result))[1]

# Вывод результата.

print(f'Результат решения № 2: Максимальный объем продаж на рекламном канале: {max_value_2}')
```

```python
# Задание 5. Словарь с вложеным словарем, в котором вложен словарь ... N

"""

Пользователь вводит произвольный список значение.
Программа делает вложеный словарь в словарь.

my_list = [1, 2, 3, 4, 5, 6]

{1: {2: {3: {4: {5: 6}}}}}
"""

# Введенные данные пользователя.

my_list = ['a', 'b', 'c', 'd', 'e', 'f']

# Объявление переменных.

new_list = dict()

# Берем последний элемент (стек - LIFO)

item_last = my_list.pop()

# Переварачиваем список и добавляем последний элемент стека в качестве значения, а первую итерацию в качестве ключа.
# Каждый последний элемент будет значением => каждая следующая итерация будет ключом.

for item in reversed(my_list):
    item_last = {item: item_last}
print(item_last)
```
