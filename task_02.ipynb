```python
# Задание 1. Средняя буква в слове.

"""

Пользователь вводит слово с клавиатуры.
Программа показывает среднюю букву в слове. Если слово четное 2 буквы.

Например:

>>>Введите слово:Москва
ск

"""

word = str(input('Ввведите слово: '))
word_len = len(word) 
word_middle = word_len // 2

if word_len % 2 == 0:
    print(word[word_middle - 1:word_middle + 1])

else:
    print(word[word_middle:word_middle + 1])
```

```python
# Задание 2. Сумма чисел.

"""

Пользователь вводит число с клавиатуры.
Программа считает сумму всех введенных чисел.
Если пользователь вводит - 0, то программа прекращает работу и выводит результат.

Например:

>>>Введите число: 1
>>>Введите число: 2
>>>Введите число: 3
>>>Введите число: 0
6

"""

sum_number = 0
number = int(input('Введите число: '))
while number != 0:
    sum_number += number
    number = int(input('Введите число: '))
print(sum_number)
```

```python
# Задание 3. Идеальный пары.

"""

Пользователь вводит списки пар.
Программа считывает отсортированные списки и показывает идеальную пару.
Если кому-то не хватит пары, то программа сообщает об этом.

Например:

boys = ['Misha', 'Alex']
girls = ['Kate', 'Liza']

Misha и Kate
Alex и Liza

"""

boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

if len(boys) != len(girls):
    print('Внимание, кто-то может остаться без пары!')
    
else:
    boys.sort()
    girls.sort()

    print('Идеальные пары:')
    for lovers_best in zip(boys, girls):
        print(f'{lovers_best[0]} и {lovers_best[1]}')
```

```python
# Задание 4. Средняя температура в странах.

"""

Пользователь вводит среднедневную температуру в Фаренгейтах по странам.
Программа рассчитает среднюю температуру за период в Цельсиях для каждой страны.

Например:

countries_temperature = [
['RUS', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
['UA', [57.2, 55.4, 59, 59, 53.6]],
]

Средняя температура в странах:
RUS - 23.9 C
UA - 13.8 C

"""

countries_temperature = [
['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
['Germany', [57.2, 55.4, 59, 59, 53.6]],
['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]

print('Средняя температура в странах: ')
for country in countries_temperature:
    temp_sum = 0
    for temp in country[1]:
        temp_sum += temp
    temp_middle = round((temp_sum / len(country[1]) - 32) * 5 / 9, 1)
    print(f'{country[0]} - {temp_middle} C')
```

```python
# Задание 5. Уникальные пользователи.

"""

Пользователь вводит логов по количеству просмотренных страниц для каждого пользователя.
Программа считает среднее значение просмотров на пользователя.

Например:

stream = [
'2018-01-01,user1,3',
'2018-04-04,user1,13',
'2018-01-05,user2,7',
'2018-06-14,user3,4',
'2018-07-02,user3,10',
'2018-03-21,user4,19',
'2018-03-22,user4,4',
]

Среднее количество просмотров на уникального пользователя: 15.0

"""

stream = [
'2018-01-01,user1,3',
'2018-01-07,user1,4',
'2018-03-29,user1,1',
'2018-04-04,user1,13',
'2018-01-05,user2,7',
'2018-06-14,user3,4',
'2018-07-02,user3,10',
'2018-03-21,user4,19',
'2018-03-22,user4,4',
'2018-04-22,user4,8',
'2018-05-03,user4,9',
'2018-05-11,user4,11',
]

# Объявление переменных

user_unique = set()
user_list = list()
total = 0
sum_looking = 0

# Список всех уникальных пользователей

for user in stream:
    user = user.split(',')
    user_list.append(user[1])

# Удалить дубликаты

user_unique = set(user_list)

# Срденее количество для всех уникальных пользователей

for user in user_unique:
    sum_looking = 0
    for user_looking in stream:
        user_looking = user_looking.split(',')
        if user_looking[1] == user:
            sum_looking += int(user_looking[2])
    total += sum_looking
    
print('Среднее количество просмотров на уникального пользователя:', round(total / len(user_unique), 2))

```

```python
# Задание 6. Повторяющиеся числа.

"""

Пользователь вводит с клавиатуры числа через пробел.
Программа считает и показывает числа, которые повторяются больше одного раза.

>>>Введите числа через пробел: 1 2 3 4 1 2 3 4 5
Результат:
 1 2 3 4

"""

# Объявление переменных 

numbers_unique = list()
numbers_input = list()

# Ввод с клавиатуры

number = input('Введите числа через пробел: ')
numbers_input = number.split(' ')

# Находим повторяющиеся значения

for numbers in numbers_input:
    if numbers not in numbers_unique:
        if numbers_input.count(numbers) > 1:
            numbers_unique.append(numbers)

print('Результат:', *numbers_unique)
```
