```python
import pandas as pd
```

```python
# Задание № 1. Фильм с большим колличеством оценок - 5

# Читаем файлы

data_movies = pd.read_csv('./task_10_data/movies.csv')
data_ratings = pd.read_csv('./task_10_data/ratings.csv')

# Соединяем таблицы и берем нужные столбцы (ID, title, rating)

joined_dm_dr = data_movies.merge(data_ratings)
joined_dm_dr[['movieId', 'title', 'rating']]

# Находим фильм с большим колличеством оценок - 5

def rating(value):
    filtered_top = joined_dm_dr[(joined_dm_dr['rating'] == value)]
    top = filtered_top.groupby(['movieId', 'title'])['movieId'].count()
    return top.sort_values(ascending=False).head()

rating(5)
```

```python
# Задание № 2. Посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония).
# Категоря: 4, 12 и 21 за период с 2005 по 2010 года. 
# Не учитывайте в расчетах отрицательные значения quantity.

# Читаем файл

data_power = pd.read_csv('./task_10_data/power.csv')

# Создаем фильтр: 

filter_country = data_power[ 
                            # страны Прибалтики 
                            ( 
                            (data_power['country'] == 'Latvia') |
                            (data_power['country'] == 'Lithuania') |
                            (data_power['country'] == 'Estonia')
                            ) &
                            # категоря: 4, 12 и 21 
                            ( 
                            (data_power['category'] == 4) |
                            (data_power['category'] == 12) |
                            (data_power['category'] == 24)
                            ) & 
                            # период с 2005 по 2010 года
                            (
                            (data_power['year'] >= 2005) & \
                            (data_power['year'] <= 2010)
                            ) &
                            # только положительные
                            (data_power['quantity'] >= 0) ]

filter_country['quantity'].sum()
```
