```python
# Перечень всех документов

documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

# Перечень полок

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

```python
# Показывает весь список полок

def shelf():
    """
    Показывает весь список полок
    
    shelf()
    
    Текущий перечень полок: 1 2 3 4 5
    """
    shelfs = [int(shelf) for shelf in directories]
    shelfs.sort()
    print('Текущий перечень полок:',*shelfs)
```

```python
# Задание 1. Пользователь по команде “p” может узнать владельца документа по его номеру.

def serch_name():
    """
    Показывает владельца по номеру документа
    
    print(serch_name())
    
    >>>Введите номер документа: 11-2
    'Геннадий Покемонов'
    """
    doc_input = input('Введите номер документа: ')
    for doc in documents:
        if doc['number'] == doc_input:
            return print(f"Владелец документа: {doc['name']}")
    return print('Документа не найден в базе')
```

```python
# Задание 2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится.

def serch_shelf():
    """
    Поиск полки по номеру документа

    print(serch_shelf())

    >>>Введите номер документа: 10006
    Документ хранится на полке: 2
    """
    doc_input = input('Введите номер документа: ')
    for shelf, number_doc in directories.items():
        if doc_input in number_doc:
            return print(f"Документ хранится на полке: {shelf}")
    return print('Документ не найден в базе')
```

```python
# Задание 3. Пользователь по команде “l” может увидеть полную информацию по всем документам.

def full_info():
    """
    Показывает информацию по всем документам
    
    full_info()
    
    №: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1
    """
    for document in documents:
        for shelf_num, doc_dir in directories.items():
            if document['number'] in doc_dir:
                number = '№: ' + document['number']
                type_ = 'тип: ' + document['type']
                name = 'владелец: ' + document['name']
                shelf_num = 'полка хранения: ' + shelf_num
                print(f'{number} {type_} {name} {shelf_num}')
```

```python
# Задание 4. Пользователь по команде “as” может добавить новую полку.

def add_shelf(shelf_input=None):
    """
    Добавляет новую полку
    
    add_shelf(shelf_input=None)
    
    >>>Введите номер полки: 4
    Полка добавлена. Текущий перечень полок: 1 2 3 4 5
    """
    if shelf_input is None:
        shelf_input = input('Введите номер полки: ').strip()
    if shelf_input.isdigit():
        if shelf_input in directories:      
            print('Такая полка уже есть. ', end = '')
            shelf()
        else:
            directories[shelf_input] = []
            print('Полка добавлена.', end = '')
            shelf()
    else:
        print('Вы выели не верное значение')
```

```python
# Задание 5. Пользователь по команде “ds” может удалить существующую полку из данных.

def del_shelf():
    """
    Удаляет полку
    
    del_shelf()
    
    >>>Введите номер полки: 2
    На полке есть документа, удалите их перед удалением полки.Текущий перечень полок: 1 2 3
    """
    shelf_input = input('Введите номер полки: ')
    for num_shelf, num_doc in directories.items():
        if shelf_input == num_shelf and not num_doc:
            directories.pop(num_shelf)
            print('Полка удалена.', end = '')
            return shelf()
        elif shelf_input == num_shelf:
            print('На полке есть документа, удалите их перед удалением полки.', end = '')
            return shelf()
    print('Такой полки нет. ', end = '')
    shelf()
```

```python
# Задание 6. Пользователь по команде “ad” может добавить новый документ в данные.

def add_doc():
    """
    Добавляет документ в базу
    
    add_doc()
    
    >>>Введите полку для хранения: 5
    >>>Такой полки нет или она уже создание. Вы хотите добавить полку ?(y/n): y
    Полка добавлена.Текущий перечень полок: 1 2 3 5
    >>>Введите номер документа: 4545
    >>>Введите тип документа: top secret
    >>>Введите владельца документа: DJ
    №: 4545 тип: top secret владелец: DJ полка хранения: 5
    """
    shelf_input = input('Введите полку для хранения: ')
    if (shelf_input not in directories) or \
        (shelf_input in directories and \
         directories[shelf_input] == []):
        change_iunput = input('Такой полки нет или она уже создание. Вы хотите добавить полку ?(y/n): ').lower()
        if change_iunput == 'y':
            add_shelf(shelf_input)
        else:
            print('Полка не добавлена. Создайте полку с помощь команды: as')
            return full_info()  
        doc_input = input('Введите номер документа: ')
        type_input = input('Введите тип документа: ')
        name_input = input('Введите владельца документа: ')

        doc_new = {'type':type_input, 'number':doc_input, 'name':name_input}

        for doc in documents:
            if (doc['type'] == doc_new['type'] and \
                doc['number'] == doc_new['number'] and \
                doc['name'] == doc_new['name']):
                print('Документ уже есть!')
                return full_info()
        documents.append(doc_new)
        directories[shelf_input].append(doc_input)
        return full_info()
    else:
        print('Полка занята')
        return full_info()
```

```python
# Задание 7. Пользователь по команде “d” может удалить документ из данных.

def del_document():
    """
    Удаляет документ с полки
    
    del_document()
    
    >>>Введите номер документа: 11-2
    Документ удален
    №: 2207 876234 тип: passport владелец: Василий Гупкин полка хранения: 1
    №: 10006 тип: insurance владелец: Аристарх Павлов полка хранения: 2
    """
    doc_input = input('Введите номер документа: ')
    for shelf, docs in directories.items():
        if doc_input in docs:
            docs.remove(doc_input)
            print('Документ удален.')
            return full_info()
    print('Документ не найден в базе')
    return full_info()
```

```python
# Задание 8. Пользователь по команде “m” может переместить документ с полки на полку.

def move_doc():
    """
    Перемещает документ
    
    move_doc()
    
    >>>Введите номер документа: 12
    >>>Введите номер полки: 32
    Такой полки не существует.Текущий перечень полок: 1 2 3
    """
    doc_input = input('Введите номер документа: ')
    shelf_input = input('Введите номер полки: ')

    for shelfs, doc in directories.items():
        if (doc_input in directories[shelfs] \
        and shelf_input in directories):
            directories[shelfs].remove(doc_input)
            directories[shelf_input].append(doc_input)
            print('Документ перемещен')
            return full_info()
    if shelf_input not in directories:
        print('Такой полки не существует.', end = '')
        return shelf()
    else:
        print('Такого документа не существует. ')
        return full_info()
```

```python
# Задание 0. Пользовательский ввод обработки в цикле while до тех пор, пока пользователь явно не завершит программу.

def main():
    """
    Пользователь вводит номера полок и комплект документов
    Программа запускается через основную функцию main()
    Более подробнее программы, при открытие основной функции.
    
    main()
    """
    while True:
        print('\nДоступеые команды:\n\
        p - узнать владельца документа по номеру\n\
        s - узнать номер полки по номеру документа\n\
        l - полная информация по всем документам\n\
        as - добавление новой полки\n\
        ds - удаление полки\n\
        ad - добавить новый документ\n\
        d - удаляет документ с полки\n\
        m - перемещает документ\n\
        q - выход')
        user_input = input('\nВведите команду: ').lower()
        if user_input == 'p':
            serch_name()
        elif user_input == 's':
            serch_shelf()
        elif user_input == 'l':
            full_info()
        elif user_input == 'as':
            add_shelf()
        elif user_input == 'ds':
            del_shelf()
        elif user_input == 'ad':
            add_doc()
        elif user_input == 'd':
            del_document()
        elif user_input == 'm':
            move_doc()
        elif user_input == 'q':
            break
        else:
            print('Данной команды нету. Повторите снова\n')
main()
```
